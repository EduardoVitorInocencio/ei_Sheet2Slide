### Acessando o menor nível dos elementos agrupados

```python
from pptx import Presentation

# Caminho do arquivo
pptx_path = r"C:\Users\70088884\beauty_sales_dashboard\templates\template_vendas_atualizado.pptx"

# Abre o PPT
ppt = Presentation(pptx_path)

# Seleciona o primeiro slide
slide = ppt.slides[0]

# Função recursiva para imprimir todos os shapes dentro de um grupo
def print_shapes_recursive(group, indent=0):
    prefix = " " * indent
    for shape in group.shapes:
        print(f"{prefix}- {shape.name} (type={shape.shape_type})")
        # Se o shape também for um grupo, percorre recursivamente
        if shape.shape_type == 6:  # 6 = msoGroup
            print_shapes_recursive(shape, indent + 4)

# Encontra Group 5
group5 = next((s for s in slide.shapes if s.shape_type == 6 and s.name == "Group 5"), None)

if group5:
    print("Conteúdo de cada subgrupo dentro do Group 5:")
    # Percorre apenas os subgrupos cujo nome começa com "Group"
    for subshape in group5.shapes:
        if subshape.shape_type == 6 and subshape.name.startswith("Group"):
            print(f"\nSubgrupo: {subshape.name}")
            print_shapes_recursive(subshape, indent=4)
else:
    print("Group 5 não encontrado no slide 1")
```

### Gerando um Json contendo o nome dos elementos do slide

Essa lista será utilizada para atualizar os objetos existentes nos slides, com base nas variáveis existentes no arquivo.

```python
import json
from pptx import Presentation

# Caminho do arquivo PPTX
pptx_path = r"C:\Users\70088884\beauty_sales_dashboard\templates\template_vendas_atualizado.pptx"

ppt = Presentation(pptx_path)
slide = ppt.slides[0]

# Função recursiva para coletar shapes por prefixo
def collect_shapes_by_prefix(group, sections):
    for shape in group.shapes:
        name_lower = shape.name.strip().lower()
        if name_lower.startswith("shape_header"):
            sections["header"].append(shape.name)
        elif name_lower.startswith("shape_abs"):
            sections["abs"].append(shape.name)
        elif name_lower.startswith("shape_var"):
            sections["var"].append(shape.name)

        # Se for grupo, percorre recursivamente
        if shape.shape_type == 6:  # msoGroup
            collect_shapes_by_prefix(shape, sections)

# Encontra Group 5
group5 = next((s for s in slide.shapes if s.shape_type == 6 and s.name == "Group 5"), None)

if group5:
    # Dicionário para armazenar os shapes por seção
    sections = {
        "header": [],
        "abs": [],
        "var": []
    }

    collect_shapes_by_prefix(group5, sections)

    # Converte em JSON
    json_output = json.dumps(sections, indent=4)
    print("Shapes organizados por seção:")
    print(json_output)
else:
    print("Group 5 não encontrado no slide 1")
```

#### **Estudo inicial dos filtros**

```python
# Filtros fixos
FILTERS = {
"formats": "T. Embalagens",
"region": "Total Brasil",
"channels": "Hipermercados",
"periods": "YTD Dec-20",
"brands": "Coca-Cola Cia"
}

# Colunas que queremos no DataFrame final
SELECT_COLS = [
"periods",
"units","value_(r$)","weighted_hholds","weighted_buyers","penetration_(%)",
"units_per_buyer","spend_per_buyer_(r$)","units_per_trip","spend_per_trip",
"frequency","avg_price_per_unit"
]

COLS_RENAME = {
"units": "Units",
"value_(r$)": "Value_(R$)",
"weighted_hholds": "Weighted_HHOLDS",
"weighted_buyers": "Weighted_Buyers",
"penetration_(%)": "Penetration_(%)",
"units_per_buyer": "Units_per_Buyer",
"spend_per_buyer_(r$)": "Spend_per_Buyer_(R$)",
"units_per_trip": "Units_per_Trip",
"spend_per_trip": "Spend_per_Trip",
"frequency": "Frequency",
"avg_price_per_unit": "Avg_Price_per_Unit"
}
```

```python
# Aqui retorna os dois períodos de análise que são informados no filtro

COLUMNS_NAME = SELECT_COLS
COLUMNS_NAME.remove('periods')

df.iloc[:,5:].columns = COLUMNS_NAME

# aplica mapeamento ignorando maiúsculas/minúsculas e garante minúsculas no final
df.columns = [COLS_RENAME.get(col.lower(), col).lower() for col in df.columns]

# Filtros ** Aqui deve ser dinâmico, pois o usuário irá passar os filtros pelo endpoint e depois passar para a função
formats = FILTERS.get('formats')
region = FILTERS.get('region')
channels = FILTERS.get('channels')
period1 = 'YTD Dec-23'
period2 = 'YTD Dec-24'
brands = 'B-Brands (Outros)'

df_filtered = df[
        (df['formats']==formats) & 
        (df['region']==region) &
        (df['channels']==channels) &
        (df['brands']==brands) &
        (df['periods'].isin([period1,period2]))
    ].copy()

# 2. Separa linha antiga e linha atual
row1 = df_filtered[df_filtered['periods']==period1].squeeze()
row2 = df_filtered[df_filtered['periods']==period2].squeeze()

# 3. Peega somente as colunas numéricas
num_cols = df_filtered.select_dtypes(include='number').columns

# 4. Calcula as variações
variations  = {}
for col in num_cols:
    if col in ['frequency','penetration_(%)']:
        variations[f'var_{col}'] = row2[col]-row1[col]
    else:
        # variações relativas (%): (novo-antigo)/antigo * 100
        variations[f'var_{col}'] = round((row2[col] - row1[col]) / row1[col]*100,2)

# 5. Cria dataframe final: linha do período atual + variações
df_final = pd.DataFrame([{**row2.to_dict(), **variations}])
```

### Função pronta para aplicar no objeto na factory para geração da árvore de medidas tradicional

```python
# Essa função recebe os dados, os parâmetros de filtros e as colunas que farão parte do chart.

def generate_variation_table(df, filters, period1, period2, select_cols, cols_rename):
    """
    Gera tabela filtrada e calcula variações entre dois períodos.
    
    Parâmetros:
    - df : pd.DataFrame original
    - filters : dict com filtros (ex: {"formats": "T. Embalagens", "region": "Total Brasil", ...})
    - period1 : str, período antigo
    - period2 : str, período recente
    - select_cols : list, colunas a selecionar
    - cols_rename : dict, mapeamento de nomes de colunas
    
    Retorna:
    - pd.DataFrame com a linha do período recente + variações
    """
    
    df = df.copy()
    
    # --- Atualiza nomes das colunas a partir da 6ª coluna
    columns_name = select_cols.copy()
    if 'periods' in columns_name:
        columns_name.remove('periods')
    df.iloc[:,5:5+len(columns_name)].columns = columns_name
    
    # --- Renomeia colunas usando cols_rename (case-insensitive) e força minúsculas
    df.columns = [cols_rename.get(col.lower(), col).lower() for col in df.columns]
    
    # --- Aplica filtros dinamicamente
    df_filtered = df.copy()
    for key, value in filters.items():
        df_filtered = df_filtered[df_filtered[key.lower()] == value]
    
    # --- Filtra apenas os dois períodos
    df_filtered = df_filtered[df_filtered['periods'].isin([period1, period2])]
    
    if df_filtered.shape[0] < 2:
        raise ValueError("Não foram encontradas linhas suficientes para os períodos especificados.")
    
    # --- Separa linha antiga e linha atual
    row1 = df_filtered[df_filtered['periods'] == period1].squeeze()
    row2 = df_filtered[df_filtered['periods'] == period2].squeeze()
    
    # --- Colunas numéricas
    num_cols = df_filtered.select_dtypes(include='number').columns
    
    # --- Calcula variações
    variations = {}
    for col in num_cols:
        if col in ['frequency', 'penetration_(%)']:
            variations[f'var_{col}'] = row2[col] - row1[col]
        else:
            variations[f'var_{col}'] = round((row2[col] - row1[col]) / row1[col] * 100, 2)
    
    # --- Cria dataframe final
    df_final = pd.DataFrame([{**row2.to_dict(), **variations}])
    
    return df_final
```
